```python
data = pd.read_csv('london_merged.csv')  # reading the csv file using pandas 
```

```python
data.info()  # to get a basic idea of the column names and their data types
```

```python
data.shape  # to get an idea of rows and columns in the table
```

```python
new_col = {'timestamp': 'time',
           'cnt': 'count',
           't1' : 'real_temp',
           't2': 'feels_like_temp',
           'hum': 'humidity_percent',
           'wind_speed': 'windspeed_kph',
           'weather_code':'weather',
           'is_holiday':'is_holiday',
           'is_weekend':'is_weekend',
           'season': 'season'}   # creating a dict with column names that I plan to use

#assigning new names to the columns
data.rename(new_col, axis=1, inplace = True)
data.head()
 
```

```python
data.head()
```

```python
# as the data contains number values for season creating a new dictonary to use str
season_ = {'0.0': 'spring',
           '1': 'summer',
           '2': 'fall',
           '3': 'winter'}

# doing the same as I did for season to weather

weather_ = {'1.0': 'clear',
            '2.0': 'scattered clouds',
            '3.0': 'broken clouds',
            '4.0':'cloudy',
            '7.0': 'rain',
            '10.0': 'rain with thunderstorm',
            '26.0': 'snowfall'}

#changing the data type of season column from float to str
data.season = data.season.astype('str')  

# mapping the values to season column
data.season = data.season.map(season_)

#doing the same for weather that I did for season just above

data.weather = data.weather.astype('str')

#mapping
data.weather = data.weather.map(weather_)

# confirming the changes
data.head()


```

```python
data.to_excel('london_bikes_final.xlsx', sheet_name= 'Bikes')
```
